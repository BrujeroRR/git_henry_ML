Ahora haremos el sentiment analysis

```python
reviews = pd.read_csv('Steam_Datasets/reviews_desanidado.csv')
reviews.drop(columns='Unnamed: 0', inplace=True)
reviews.head()
```

Instalamos e importamos la biblioteca TextBlob

```python
from textblob import TextBlob
```

Ahora para el sentiment analysis usaremos la función polarity

```python
reviews['reviews'] = reviews['reviews'].astype(str)

# La funcion polarity se utiliza para calcular la polaridad del sentimiento de un texto
def calcular_polaridad(reviews):
    blob = TextBlob(reviews)
    return blob.sentiment.polarity

# Para crear la nueva columna Polarity tuvimos que usar la funcion en la columna review

reviews['polarity'] = reviews['reviews'].apply(calcular_polaridad)


reviews.head()
```

Asignamos los valores 0, 1 y 2 dependiendo de la polaridad de la review

```python
def colocar_valores(polaridad):
    if polaridad < -0.2:
        return 0  # Sin sentimiento
    elif -0.2 <= polaridad <= 0.2:
        return 1  # Sentimiento neutral
    else:
        return 2  # Sentimiento positivo

reviews['sentiment_value'] = reviews['polarity'].apply(colocar_valores)
```

```python
reviews.head()
```

Guardamos el dataset sin la columna review para facilitarlo al modelo de ML

```python
sentiment_analysis = reviews[['item_id', 'user_id', 'recommend', 'polarity', 'sentiment_value']]
sentiment_analysis.to_csv('Steam_Datasets/reviews_sentiment.csv', index=False)
```

# Dataset items

```python
items = pd.read_csv('Steam_Datasets/Items.csv')
```

```python
items.head()
```

Borramos las columnas innecesarias

```python
items.drop(columns=['Unnamed: 0', 'user_url'], inplace=True)
```

```python
items.head()
```

Convertimos los valores de la columna items en lista

```python

def convertir_a_lista(valor):
    if isinstance(valor, str):
        try:
            return ast.literal_eval(valor)
        except (ValueError, SyntaxError):
            
            return valor
    else:
        return valor

items['items'] = items['items'].apply(convertir_a_lista)

items.head()
```

Desanidamos la lista de items

```python
items_desanidado = items.explode('items')

items_desanidado.head()
```

Borramos los registros nulos en items

```python
items_desanidado.dropna(subset='items', inplace=True)
```

Guardamos en archivo csv

```python
items_desanidado.to_csv('Steam_Datasets/items_desanidado.csv', index=False)
```
